```python
import cymetric as cym
import pandas as pd
import numpy as np
from pyne import nucname
```

```python
db = cym.dbopen('cyclus.sqlite')

def get_transaction_TS(db, fac, *args):

    nuc_list = []
    for inx, nuc in enumerate(args):
        nuc_list.append(nucname.id(nuc))
        

    print(nuc_list)
    evaler = cym.Evaluator(db)
    inv = evaler.eval('ExplicitInventory')
    #return inv

#    df_nuc_lif = pd.Series(nuc_list, name='NucId')
    selected_inv = inv.loc[inv['NucId'].isin(nuc_list)]
    #selected_inv = inv.loc[lambda df: df.NucId == df_nuc_lif,:]
    return selected_inv
```
