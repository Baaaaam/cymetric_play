```python
import cymetric as cym
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#load db in a evaler
with cym.dbopen('cyclus_case4.sqlite') as db:
    evaler = cym.Evaluator(db)
```

```python
db.tables
```

```python
#get transation & Agent tables
inv = evaler.eval('ExplicitInventory')
inv
```

```python
agents = evaler.eval('AgentEntry')
agents
```

```python
#Merge AgentId into inventory
df = pd.merge(agents[['SimId', 'AgentId', 'Prototype']], inv, on=['SimId', 'AgentId'])
df = df.drop('AgentId',1)
#df
```

```python
inv_tot = df[['Prototype', 'Time','Quantity']].groupby(['Prototype', 'Time']).sum()
#inv_tot
```

```python
inv_nuc = df[['Prototype', 'NucId', 'Time','Quantity']].groupby(['Prototype', 'NucId', 'Time']).sum()
inv_nuc
```

```python
toplot = inv_tot.loc['Storage_E1_second']
toplot.reset_index(inplace=True)
toplot
```

```python
myplot = pd.DataFrame({'Time': toplot.Time/12., 'inv': toplot.Quantity})
myplot = myplot.groupby('Time').sum()
myplot.plot(kind='area', figsize=(10, 6))
plt.xlabel('Time [Years]')
plt.ylabel('Transaction [Kg]')
plt.savefig('test.png', dpi=326)
```
